```python
class ATM:
    def __init__(self, bank_name, account_name, account_number, balance,current_pin=67890):
        self.bank_name = bank_name
        self.account_name = account_name
        self.account_number = account_number
        self.balance = balance
        self.current_pin=current_pin
    def welcome_msg(self):
        print(f"******************************Welcome to {self.bank_name} ATM*************************************")
        print("\n\n                              Insert your card                                             ")

    def validate_pin(self, pin_attempts=0):
        entered_pin = int(input("Enter the pin: "))
        if entered_pin == self.current_pin:
            print("Your pin is correct")
            return True
        else:
            print("Please enter correct pin")
            pin_attempts += 1
            if pin_attempts >= 3:  
                print("Too many incorrect attempts. Account locked.")
                return False
            else:
                return self.validate_pin(self.current_pin, pin_attempts)

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient amount")
        else:
            self.balance -= amount
            print(f"You withdraw the amount: {amount} and the current balance is {self.balance}")

    def fast_withdraw(self):
        amount = int(input("Enter the amount to fast withdraw (2000 or 4000): "))
        if amount in [2000, 4000]:
            if amount > self.balance:
                print("Insufficient balance")
            else:
                self.balance -= amount
                print(f"You withdraw the amount: {amount} and the current balance is {self.balance}")
        else:
            print("Invalid input")

    def quick_withdraw(self):
        amount = int(input("Enter the amount to quick withdraw (200, 400, 600, 800, 1000, 3000): "))
        if amount in [200, 400, 600, 800, 1000, 3000]:
            if amount > self.balance:
                print("Insufficient balance")
            else:
                self.balance -= amount
                print(f"You withdraw the amount: {amount} and the current balance is {self.balance}")
        else:
            print("Invalid input")

    def deposit(self, amount):
        self.balance += amount
        print(f"You deposit amount: {amount} and the balance is {self.balance}")

    def check_balance(self):
        return self.balance

    def change_pin(self, new_pin, re_enter_pin):
        if self.current_pin == int(input("Enter your current pin:")):
            if new_pin == re_enter_pin:  
                self.current_pin = new_pin
                print("Pin changed successfully")
            else:
                print("Pins do not match. Please try again.")
        else:
            print("Current pin is incorrect.")

    def receipt(self):
        print(f"Bank name: {self.bank_name}\nAccount number: {self.account_number}\nAccount name: {self.account_name}\nBalance: {self.balance}")

def main():          
    atm = ATM("Canara Bank", "Moni", 1245788, 9000)  
    atm.welcome_msg()
    if not atm.validate_pin():
        return  
    while True:
        print("\n1. Deposit\n2. Withdraw\n3. Fast Withdraw \n4. Quick withdraw \n5. Check Balance\n6. Change PIN\n7. Receipt Generation\n8. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            amount = float(input("Please enter the amount to deposit: "))
            atm.deposit(amount)
        elif choice == "2":
            amount = float(input("Please enter the amount to withdraw: "))
            atm.withdraw(amount)
        elif choice == "3":
            atm.fast_withdraw()
        elif choice == "4":
            atm.quick_withdraw()
        elif choice == "5":
            print("Your balance is:", atm.check_balance())
        elif choice == "6":
            new_pin = input("Enter new pin: ")
            re_enter_pin = input("Re-enter the pin: ")
            atm.change_pin(new_pin, re_enter_pin)
        elif choice == "7":
            atm.receipt()
        elif choice == "8":
            print("Thank you for using our ATM. Goodbye!")
            break
        else:
            print("Invalid choice")
       
        while True:
            continue_choice = input("\nDo you want to continue? (1. Yes / 2. No): ")
            if continue_choice == "1":
                break
            elif continue_choice == "2":
                print("Thank you for using our ATM. Goodbye!")
                return
            else:
                print("Invalid choice. Please enter '1' for Yes or '2' for No.")

if __name__ == "__main__":
    main()
```
